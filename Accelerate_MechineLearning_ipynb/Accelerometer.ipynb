# 三轴加速度动作识别

```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
df_data1 = pd.read_csv(r'UpStairsAccelerometer.csv')#数据使用“SensorRecord”采集android三轴加速度传感器信息
del df_data1['Milliseconds']
df_data1_5 = pd.read_csv(r'1.5LeftUpStairsAccelerometer.csv')
del df_data1_5['Milliseconds']
df_data2 = pd.read_csv(r'DownStairsAccelerometer.csv')
del df_data2['Milliseconds']
df_data3 = pd.read_csv(r'FlatWalkAccelerometer.csv')
del df_data3['Milliseconds']
df_data01 = pd.read_csv(r'StandAccelerometer.csv')
del df_data01['Milliseconds']
df_data02 = pd.read_csv(r'UpElevatorAccelerometer.csv')
del df_data02['Milliseconds']
df_data03 = pd.read_csv(r'DownElevatorAccelerometer.csv')
del df_data03['Milliseconds']
df_data=pd.concat([df_data1,df_data1_5,df_data2,df_data3,df_data01,df_data02,df_data03])

df_label1 = pd.read_csv(r'UpStairsAccelerometerLabel.csv')
df_label1_5 = pd.read_csv(r'1.5LeftUpStairsAccelerometerLabel.csv')
df_label2 = pd.read_csv(r'DownStairsAccelerometerLabel.csv')
df_label3 = pd.read_csv(r'FlatWalkAccelerometerLabel.csv')
df_label01 = pd.read_csv(r'StandAccelerometerLabel.csv')
df_label02 = pd.read_csv(r'UpElevatorAccelerometerLabel.csv')
df_label03 = pd.read_csv(r'DownElevatorAccelerometerLabel.csv')
df_label=pd.concat([df_label1,df_label1_5,df_label2,df_label3,df_label01,df_label02,df_label03])
```

```python
df_label02.values.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_data, df_label, test_size=0.1)
```

```python
outputpath='total.csv'#其中：index是否要索引，header是否要列名，True就是需要
df_data.to_csv(outputpath,sep=',',index=False,header=False)

outputpath='totalLabel.csv'
df_label.to_csv(outputpath,sep=',',index=False,header=False)
```

```python
outputpath='x_train.csv'#其中：index是否要索引，header是否要列名，True就是需要
x_train.to_csv(outputpath,sep=',',index=False,header=False)

outputpath=' x_test.csv'
x_test.to_csv(outputpath,sep=',',index=False,header=False)

outputpath='y_train.csv'#其中：index是否要索引，header是否要列名，True就是需要
y_train.to_csv(outputpath,sep=',',index=False,header=False)

outputpath=' y_test.csv'
y_test.to_csv(outputpath,sep=',',index=False,header=False)
```

```python
x_train=x_train.values
y_train=y_train.values
x_test=x_test.values
y_test=y_test.values
```

```python
import time
time_start=time.time()
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, input_dim=3, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(4, activation='softmax')
])
#输入层3维样本X，Y，Z，神经元分别个不同的层，然后随机失活dropout=0.2,最后计算结果
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
#WARNING:tensorflow:Falling back from v2 loop because of error: Failed to find data adapter that can handle input: <class 'pandas.core.frame.DataFrame'>, <class 'NoneType'>
#pd.DataFrame2Numpy
#优化器使用Adam优化算法，交叉熵损失函数，学习率0.001，精度评价模型
model.fit(x_train, y_train, batch_size=4, epochs=9)
#训练
time_end=time.time()
print('time cost',time_end-time_start,'s')
```
